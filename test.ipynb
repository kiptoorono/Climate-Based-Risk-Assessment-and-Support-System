```python
import pandas as pd

# Load your data (replace 'data.xlsx' and 'Sheet1' with actual file/sheet names)
df = pd.read_excel("E:/Agriculture project/Data/Semi-prepared data/Rainfall_total.xlsx", sheet_name="Sheet1", index_col=0)
# df = pd.read_excel('E:/Agriculture project/Data/Semi-prepared data/Rainfall_total.xlsx', parse_dates=['Date'], index_col='Date')

# List of counties in Agro-Ecological Zones 1-4
selected_counties = [
    "Nyeri", "Meru", "Kirinyaga", "Embu", "Murang'a", "Elgeyo-Marakwet",
    "Kericho", "Bomet", "Kiambu", "Nandi", "Nyandarua", "Kisii",
    "Trans Nzoia", "Uasin Gishu", "Bungoma", "Kakamega", "Vihiga",
    "Nakuru", "Laikipia", "Kitui", "Machakos", "Makueni", "Tharaka-Nithi",
    "West Pokot", "Narok", "Baringo"
]

# Filter dataset to include only selected counties
df_filtered = df.loc[df.index.isin(selected_counties)]

# Compute per-county thresholds
low_thresholds = df_filtered.quantile(0.05)  # 5th percentile for drought
high_thresholds = df_filtered.quantile(0.95)  # 95th percentile for heavy rain

# Identify drought and heavy rain months
drought_mask = df_filtered.lt(low_thresholds, axis=1)  # Less than low threshold
heavy_rain_mask = df_filtered.gt(high_thresholds, axis=1)  # Greater than high threshold

# Count drought and heavy rain months per county
drought_months = drought_mask.sum(axis=1)
heavy_rain_months = heavy_rain_mask.sum(axis=1)

# Create results DataFrame
filtered_results = pd.DataFrame({
    'Drought Months': drought_months,
    'Heavy Rain Months': heavy_rain_months
})

# Save to Excel
filtered_results.to_excel("filtered_rainfall_analysis.xlsx", sheet_name="Drought & Heavy Rain")

print("Filtered data saved to 'filtered_rainfall_analysis.xlsx'")
```
